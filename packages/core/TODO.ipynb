```typescript
import { Builder as XMLBuilder } from 'xml2js';

// Define the data structure for the playlist
const playlistData = {
  playlist: {
    $: {
      version: "1",
      xmlns: "http://xspf.org/ns/0/",
      "xmlns:tvg": "http://example.com/tvg",
    },
    title: "TV Playlist",
    trackList: {
      track: [
        {
          title: "Channel 1",
          location: new URL("http://stream.example.com/channel1.m3u8"),
          extension: {
            $: { application: "http://example.com/tvg" },
            "tvg:tvg-id": "channel1",
            "tvg:xmltv-url": "http://xmltv.example.com/epg.xml",
          },
        },
        {
          title: "Channel 2",
          location: "http://stream.example.com/channel2.m3u8",
          extension: {
            $: { application: "http://example.com/tvg" },
            "tvg:tvg-id": "channel2",
            "tvg:xmltv-url": "http://xmltv.example.com/epg.xml",
          },
        },
      ],
    },
  },
};

// Create an XML Builder
const builder = new XMLBuilder({
  xmldec: {
    version: "1.0",
    encoding: "UTF-8",
  },
  renderOpts: {
    pretty: true, // Format with indentation
  },
});



// Generate the XML string
const xmlString = builder.buildObject(playlistData);

// Output the result
console.log(xmlString);
```
